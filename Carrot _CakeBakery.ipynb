```python
# UK Carrot Cake Bakery Business Analysis
# Jupyter Notebook Outline

# --- IMPORTS ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- BUSINESS PROFILES (Example Static Data) ---
# Two hypothetical or example bakery businesses selling carrot cake in UK

businesses = {
    "Paul Bakery": {
        "location": "UK (nationwide bakery chain with shops & online)", 
        "menu_items": ["Carrot Cake (£20)", "Coffee (£3)", "Scones (£5)"],
        "pricing": {"Carrot Cake":20, "Coffee":3, "Scones":5},
        "carrot_cake_servings": [4,6,8],
        "sales_channels": ["In-shop", "Online Delivery", "Click & Collect"]
    },
    "Bakery Store UK": {
        "location": "UK (artisan bakery sites & online)", 
        "menu_items": ["Carrot Cake Gateaux (£?)", "Pastries (£?)", "Cookies (£?)"],
        "pricing": {"Carrot Cake":22, "Pastries":5, "Cookies":3},
        "carrot_cake_servings": ["Gateaux whole cake"],
        "sales_channels": ["In-shop", "Website Orders"]
    }
}

# Convert to DataFrame
business_df = pd.DataFrame(businesses).T
business_df

# Observations (replace with real scraped/collected menu data)
# e.g. PAUL offers carrot cake at ~£20 for a cake serving 4–8; other items include coffee & scones. 
# The Bakery Store UK sells artisan carrot cake gateaux portion(s). 

# --- MENU COMPARISON ---
# Create comparison DataFrame
menu_comparison = pd.DataFrame({
    "Item": ["Carrot Cake","Coffee","Pastries","Cookies","Scones"],
    "Paul Bakery": [20,3,np.nan,np.nan,5],
    "Bakery Store UK": [22,np.nan,5,3,np.nan]
})
menu_comparison

# Visual compare pricing
menu_comparison.set_index("Item").plot(kind="bar")
plt.title("Menu Item Pricing Comparison")
plt.ylabel("Price (£)")
plt.show()

# --- DEMOGRAPHICS ASSUMPTIONS ---
# Typical customer profiles for UK bakery demand (broad industry guidance)
customer_demographics = pd.DataFrame({
    "Age Group": ["18-24","25-34","35-49","50-64","65+"],
    "Estimated % of Sales": [15,30,25,20,10],
    "Preference": ["Trendy/variety","Frequent buyers","Premium quality","Traditional tastes","Occasional purchases"]
})

customer_demographics

# Notes:
# - Younger customers (Gen Z and Millennials) are influenced by social media food trends. 
# - Carrot cake can appeal broadly to café visitors, especially during brunch/afternoon tea.

# --- SALES & PROFIT FORECAST (SIMPLIFIED MODEL) ---
# Assumptions:
# - Average weekly carrot cake units sold
# - Profit margin assumptions (bakery industry typical margins ~30-50%, varies widely)

# Define assumed weekly sales for carrot cake
weekly_sales_assumptions = {
    "Paul Bakery": {"cakes_sold": 150, "avg_price": 20, "cost_per_unit": 10},
    "Bakery Store UK": {"cakes_sold": 120, "avg_price": 22, "cost_per_unit": 11}
}

# Compute weekly revenue & profit
profit_data = []
for name, data in weekly_sales_assumptions.items():
    revenue = data["cakes_sold"] * data["avg_price"]
    profit = (data["avg_price"] - data["cost_per_unit"]) * data["cakes_sold"]
    profit_data.append({
        "Business": name,
        "Weekly Revenue (£)": revenue,
        "Weekly Profit (£)": profit,
        "Annual Revenue (£)": revenue * 52,
        "Annual Profit (£)": profit * 52
    })

profit_df = pd.DataFrame(profit_data)
profit_df

# Visualise revenue vs profit
sns.barplot(x="Business", y="Weekly Profit (£)", data=profit_df)
plt.title("Weekly Profit Comparison")
plt.show()

# --- CONCLUSIONS & STRATEGIC INSIGHTS ---
print("""
1) Pricing Strategy:
   • Paul Bakery’s carrot cake pricing is competitive for standard servings.
   • Bakery Store UK’s cake is slightly higher priced, possibly indicating artisan branding.

2) Customer Demographics:
   • Younger customers (18-34) are significant spenders on trendy bakery products due to social media influence. 
   • Older demographics contribute to consistent classic cake sales.

3) Sales & Profit Projections:
   • With assumed sales rates, both businesses show healthy weekly and yearly profits.
   • Actual performance will vary based on footfall, online marketing, seasonal demand, and cost control.

4) Market Context:
   • The UK bakery sector is growing, with industry revenue forecast to expand (over £3.7bn). 
   • Competition includes national chains (e.g., Greggs with high volumes but differing product focus) and franchise cake businesses like Cake Box. 

""")
```
